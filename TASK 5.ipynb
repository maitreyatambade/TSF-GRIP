```python

import pandas as pd
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

```

# Importing the DATA SET

```python
Sales_df= pd.read_csv('SampleSuperstore.csv') 
Sales_df
```

# Describing the Data Set

```python
Sales_df.describe()
```

```python
Sales_df.drop_duplicates(keep= 'first', inplace= True)#Dropping Duplicates 
Sales_df
```

```python
Col= Sales_df[['Ship Mode','Segment','State','Region','Category']]
```

# Determining Unique Values 

```python
for Col_Name in Col:
    print(Col_Name)
    Col = [Sales_df[Col_Name].unique()]
    for i in Col:
        print(str(i))
```

# Visualization 


```python
df= Sales_df[['Ship Mode', 'Sales', 'Quantity', 'Discount', 'Profit']]
```

```python
sns.pairplot(df, hue = 'Ship Mode')
```

## The above Graphical represents how different shipping modes affect the different aspects of the business?

```python
plt.figure(figsize= (15,9))
Sales_df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

## The above graphical figure represents the totel sales and the total profit  generated in different segments.

```python
plt.figure(figsize= (15,9))
sns.countplot('Sub-Category', hue='Region', data=Sales_df)
plt.title('Category wise Sales Count of each Region', size=30)
plt.xticks(rotation=90)
plt.show()
```
